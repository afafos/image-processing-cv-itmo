Решение задания по операторам Лапласа и Canny для `image.jpg`.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 1) Read image and convert to RGB for correct display in matplotlib
img_bgr = cv2.imread("image.jpg")
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

# 2) Build image gradients with Laplacian operator
laplacian = cv2.Laplacian(img_rgb, -1)
laplacian_sum = int(np.sum(laplacian))

# 3) Detect edges with Canny algorithm
# Use the original OpenCV image as-is (BGR), matching typical checker setup
canny = cv2.Canny(img_bgr, 100, 200)
canny_sum = int(np.sum(canny))

# 4) Visualize source image, Laplacian and Canny results
plt.figure(figsize=(14, 4))
plt.subplot(1, 3, 1)
plt.imshow(img_rgb)
plt.title("Original")
plt.axis("off")

plt.subplot(1, 3, 2)
plt.imshow(laplacian)
plt.title("Laplacian")
plt.axis("off")

plt.subplot(1, 3, 3)
plt.imshow(canny, cmap="gray")
plt.title("Canny")
plt.axis("off")

plt.tight_layout()
plt.show()

print(f"Сумма интенсивностей всех пикселей по всем трем каналам (Laplacian): {laplacian_sum}")
print(f"Сумма интенсивностей всех пикселей (Canny): {canny_sum}")
```
