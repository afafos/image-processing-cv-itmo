# Task 4.1
Решение задания по дискретному преобразованию Фурье для `image.jpg`.

```python
import cv2
import numpy as np

# 1) Read and convert image to grayscale
img = cv2.imread("image.jpg")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# 2) 2D FFT and complex component at [149, 403]
f = np.fft.fft2(gray)
z = f[149, 403]

# 3) Shift zero-frequency component to center and compute amplitudes
fshift = np.fft.fftshift(f)
amp = np.abs(fshift)

# 4) Log scaling and dynamic range
log_amp = np.log1p(amp)
log_range = log_amp.max() - log_amp.min()

# 5) Zero low frequencies in central 80x80 region and inverse transform
fshift_cut = fshift.copy()
h, w = fshift_cut.shape
cy, cx = h // 2, w // 2
fshift_cut[cy - 40: cy + 40, cx - 40: cx + 40] = 0

ishift = np.fft.ifftshift(fshift_cut)
img_back = np.fft.ifft2(ishift)
img_back_amp = np.abs(img_back)

# Print answers with precision up to hundredths
print(round(z.real, 2))
print(round(z.imag, 2))
print(round(amp.min(), 2))
print(round(amp.max(), 2))
print(round(log_range, 2))
print(round(img_back_amp[669, 428], 2))
```
