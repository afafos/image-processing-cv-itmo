# Task 14 - Карта глубины

Расчет выполнен напрямую для пары `image_left.png` / `image_right.png` методами StereoBM и StereoSGBM с параметрами из примера.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
imgL = cv2.imread('image_left.png', cv2.IMREAD_GRAYSCALE)
imgR = cv2.imread('image_right.png', cv2.IMREAD_GRAYSCALE)

if imgL is None or imgR is None:
    raise FileNotFoundError('Не удалось загрузить image_left.png или image_right.png')

plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.imshow(imgL, cmap='gray')
plt.title('Left')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(imgR, cmap='gray')
plt.title('Right')
plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
stereo_bm = cv2.StereoBM_create(numDisparities=32, blockSize=15)
disparity_BM = stereo_bm.compute(imgL, imgR)

plt.figure(figsize=(8, 4))
plt.imshow(disparity_BM, aspect='equal', cmap='jet')
plt.colorbar()
plt.title('Disparity (StereoBM)')
plt.axis('off')
plt.show()
```

```python
arr = disparity_BM.flatten()
bm_far = np.count_nonzero(arr == arr.min()) / arr.size
bm_close = np.count_nonzero(arr == arr.max()) / arr.size

print('StereoBM exact:', bm_far, bm_close)
print(f'StereoBM rounded(3): {bm_far:.3f}, {bm_close:.3f}')
print('StereoBM truncated(3):', np.floor(bm_far * 1000) / 1000, np.floor(bm_close * 1000) / 1000)
```

```python
stereo_sgbm = cv2.StereoSGBM_create(
    minDisparity=-16,
    numDisparities=32,
    blockSize=15,
    uniquenessRatio=5,
    speckleWindowSize=5,
    speckleRange=5,
    disp12MaxDiff=2,
    P1=8 * 3 * 15 ** 2,
    P2=32 * 3 * 15 ** 2,
)
disparity_SGBM = stereo_sgbm.compute(imgL, imgR)

plt.figure(figsize=(8, 4))
plt.imshow(disparity_SGBM, aspect='equal', cmap='jet')
plt.colorbar()
plt.title('Disparity (StereoSGBM)')
plt.axis('off')
plt.show()
```

```python
arr = disparity_SGBM.flatten()
sgbm_far = np.count_nonzero(arr == arr.min()) / arr.size
sgbm_close = np.count_nonzero(arr == arr.max()) / arr.size

print('StereoSGBM exact:', sgbm_far, sgbm_close)
print(f'StereoSGBM rounded(3): {sgbm_far:.3f}, {sgbm_close:.3f}')
print('StereoSGBM truncated(3):', np.floor(sgbm_far * 1000) / 1000, np.floor(sgbm_close * 1000) / 1000)
```
