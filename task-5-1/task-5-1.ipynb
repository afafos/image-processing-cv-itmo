Решение задания по вейвлет-преобразованию Хаара для `image.jpg`.

```python
import cv2
import numpy as np
import pywt
import matplotlib.pyplot as plt

# 1-2) Чтение изображения и перевод в градации серого
image_bgr = cv2.imread('image.jpg')
image_gray = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2GRAY)

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB))
plt.title('Исходное изображение')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(image_gray, cmap='gray')
plt.title('Градации серого')
plt.axis('off')
plt.show()
```

```python
# 3) DWT2 (Haar): cA, cH, cV, cD
cA, (cH, cV, cD) = pywt.dwt2(image_gray, 'haar')

# Гистограмма набора cV и максимум h(n)
hist, bins = np.histogram(cV.flatten(), bins=256)
max_hn = int(np.max(hist))
print('Максимальное значение функции h(n):', max_hn)
```

```python
# Визуализация коэффициентов (для деталей используем модуль для наглядности)\n
coeffs_to_show = [
    ('cA', cA),
    ('|cH|', np.abs(cH)),
    ('|cV|', np.abs(cV)),
    ('|cD|', np.abs(cD)),
]

plt.figure(figsize=(12, 8))
for i, (name, coeff) in enumerate(coeffs_to_show, start=1):
    plt.subplot(2, 2, i)
    plt.imshow(coeff, cmap='gray')
    plt.title(name)
    plt.axis('off')

plt.tight_layout()
plt.show()

# Ответы для скрина (изображения сверху вниз):
answers_in_screen_order = ['cH', 'cA', 'cV', 'cD']
print('Ответы для полей со скрина (сверху вниз):', ', '.join(answers_in_screen_order))
```
