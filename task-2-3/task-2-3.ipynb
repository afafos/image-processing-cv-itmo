Считываем `image.jpg`, сжимаем до ширины 720 с сохранением пропорций, переводим в RGB и выводим требуемые значения.

```python
import cv2
import os

img_bgr = cv2.imread("image.jpg")
if img_bgr is None:
    raise FileNotFoundError("Не удалось открыть image.jpg")

height, width = img_bgr.shape[:2]
ratio = width / height

new_width = 720
new_height = int(new_width / ratio)

img_resized_bgr = cv2.resize(
    img_bgr,
    (new_width, new_height),
    interpolation=cv2.INTER_AREA,
)

img_resized_rgb = cv2.cvtColor(img_resized_bgr, cv2.COLOR_BGR2RGB)

output_path = "resized_rgb.jpg"
cv2.imwrite(output_path, cv2.cvtColor(img_resized_rgb, cv2.COLOR_RGB2BGR))

print(round(ratio, 3))
print(new_height)
print(os.path.getsize(output_path))
```
