Решение задания со SIFT для `image.jpg` из текущей папки.

```python
import cv2
from pathlib import Path

image_path = Path("image.jpg")
img = cv2.imread(str(image_path))

if img is None:
    raise FileNotFoundError(f"Не удалось прочитать изображение: {image_path.resolve()}")

crop = img[217:344, 589:694]

width = crop.shape[1]
height = crop.shape[0]

print(f"Ширина кадрированного изображения: {width}")
print(f"Высота кадрированного изображения: {height}")

gray_full = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray_crop = cv2.cvtColor(crop, cv2.COLOR_BGR2GRAY)

sift = cv2.SIFT_create()
kp1, des1 = sift.detectAndCompute(gray_crop, None)
kp2, des2 = sift.detectAndCompute(gray_full, None)

bf = cv2.BFMatcher(cv2.NORM_L2, crossCheck=True)
matches = bf.match(des1, des2)

matches = sorted(matches, key=lambda m: m.distance)
best = matches[0]

x, y = kp2[best.trainIdx].pt
x_rounded = round(x)
y_rounded = round(y)

print(f"Координата X нулевой ключевой точки: {x_rounded}")
print(f"Координата Y нулевой ключевой точки: {y_rounded}")

b, g, r = img[y_rounded, x_rounded]

print(f"Интенсивность канала B: {int(b)}")
print(f"Интенсивность канала G: {int(g)}")
print(f"Интенсивность канала R: {int(r)}")
```
