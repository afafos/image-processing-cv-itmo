```python
import cv2
import matplotlib.pyplot as plt
```

```python
img_bgr = cv2.imread('image.jpg')
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10, 6))
plt.imshow(img_rgb)
plt.axis('off')
plt.show()
```

```python
# Индексы во входном условии включительные:
# строки 1393..2242 и столбцы 2420..3099
crop = img_rgb[1393:1825, 2064:2773]

plt.figure(figsize=(6, 6))
plt.imshow(crop)
plt.axis('off')
plt.show()
```
