```python
import cv2, numpy as np, pywt, urllib.request
url="http://storage.yandexcloud.net/lms-itmo-ru-files-27a87tyf/image_processing/final_task/input/aaron-burden-xtIYGB0KEqc-unsplash.jpg"
new_w=2304
row,col=891,1146
threshold=90

decode_flag=cv2.IMREAD_COLOR
interpolation=cv2.INTER_AREA
hist_bins=256
density=True
channels=3
norm_max=255
gray_mode=cv2.COLOR_BGR2GRAY
wavelet="haar"
ratio_digits=3
density_digits=4

b=urllib.request.urlopen(url).read()
i=cv2.imdecode(np.frombuffer(b,np.uint8),decode_flag)

k=i.shape[1]/i.shape[0]
h=int(new_w/k)

r=cv2.resize(i,(new_w,h),interpolation=interpolation)
m=max(np.histogram(r[:,:,c],hist_bins,density=density)[0].max() for c in range(channels))
n=np.dstack([((r[:,:,c]-r[:,:,c].min())/(r[:,:,c].max()-r[:,:,c].min())*norm_max).astype(np.uint8) for c in range(channels)])
p=n[row,col]
g=cv2.cvtColor(n,gray_mode)
ca,_=pywt.dwt2(g,wavelet)
ca1,_=pywt.dwt2(ca,wavelet)
z=int((np.abs(ca1)<threshold).sum())

print(f"{k:.{ratio_digits}f}",h,f"{m:.{density_digits}f}",int(p[0]),int(p[1]),int(p[2]),z)
```
