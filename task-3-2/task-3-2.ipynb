# Task 3.2

1. Считать `image.jpg` с помощью OpenCV.
2. Построить гистограммы интенсивностей в палитре RGB.
3. Сравнить плотности распределения по каналам R, G и B.
4. Вывести наибольшее значение плотности среди трех каналов с точностью до 4 знаков после запятой.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread("image.jpg")
if image is None:
    raise FileNotFoundError("Не удалось открыть image.jpg")

image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Повторяем логику из образца: bins=256, density=True без явного range.
hist_r, _ = np.histogram(image_rgb[:, :, 0].ravel(), bins=256, density=True)
hist_g, _ = np.histogram(image_rgb[:, :, 1].ravel(), bins=256, density=True)
hist_b, _ = np.histogram(image_rgb[:, :, 2].ravel(), bins=256, density=True)

max_density = max(hist_r.max(), hist_g.max(), hist_b.max())

plt.figure(figsize=(10, 5))
plt.plot(hist_r, color="red", label="R")
plt.plot(hist_g, color="green", label="G")
plt.plot(hist_b, color="blue", label="B")
plt.title("RGB Intensity Histograms (density)")
plt.xlabel("Intensity")
plt.ylabel("Density")
plt.xlim([0, 255])
plt.grid(alpha=0.25)
plt.legend()
plt.show()

print(f"Наибольшее значение плотности среди трех каналов: {max_density:.4f}")
```
