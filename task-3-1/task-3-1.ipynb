# Task 3.1

1. Считываем `image.jpg` с помощью OpenCV.
2. Переводим изображение в градации серого и строим гистограмму яркости.
3. Находим:
   - максимальное значение функции `h(n)`;
   - интенсивность `n`, при которой `h(n)` максимально.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread("image.jpg")
if img is None:
    raise FileNotFoundError("Не удалось открыть image.jpg")

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

hist = cv2.calcHist([gray], [0], None, [256], [0, 256]).flatten().astype(int)

h_max = int(hist.max())
n_max = int(np.argmax(hist))

plt.figure(figsize=(10, 4))
plt.title("Histogram of grayscale image")
plt.xlabel("Intensity n")
plt.ylabel("h(n)")
plt.plot(hist, color="black")
plt.xlim([0, 255])
plt.grid(alpha=0.25)
plt.show()

print("h_max =", h_max)
print("n_max =", n_max)
```
